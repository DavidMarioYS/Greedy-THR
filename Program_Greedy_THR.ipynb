```python
# Program Greedy THR dengan Konversi Mata Uang Asing

# Fungsi untuk mengonversi jumlah uang asing ke Rupiah
def convert_to_rupiah(amount, exchange_rate):
    rupiah_amount = int(amount * exchange_rate)
    return rupiah_amount


# Fungsi untuk menghitung pembagian THR menggunakan konsep greedy
def calculate_change(amount, denominations):
    result = {}
    remaining_amount = amount

    # Melakukan iterasi pada denominasi dalam urutan terurut secara menurun
    for denom in sorted(denominations, reverse=True):
        # Menghitung jumlah uang yang dibutuhkan dari denominasi saat ini
        count = remaining_amount // denom
        if count > 0:
            # Menyimpan jumlah uang dari denominasi saat ini ke dalam hasil
            result[denom] = count
            # Mengurangi jumlah uang yang belum dibagi dengan denominasi saat ini
            remaining_amount -= count * denom

    return result, remaining_amount


def main():
    print("Pilihan Mata Uang Asing:")
    print("1. Dolar Amerika Serikat (USD)")
    print("2. Euro (EUR)")
    print("3. Yen Jepang (JPY)")
    print("4. Poundsterling Inggris (GBP)")
    print("5. Dolar Australia (AUD)")
    print("6. Dolar Singapura (SGD)")
    print("7. Ringgit Malaysia (MYR)")
    print("8. Rupiah Indonesia (IDR)")

    # Meminta pengguna memilih mata uang asing
    currency_choice = int(input("Pilih mata uang asing (1-8): "))

    # Menentukan kurs mata uang asing ke Rupiah
    exchange_rates = {
        1: 14490.09,
        2: 15347.9,
        3: 113.2569,
        4: 18177.82,
        5: 10363.31,
        6: 10503.11,
        7: 3324.94,
        8: 1
    }

    # Memeriksa apakah pilihan mata uang asing valid
    if currency_choice not in exchange_rates:
        print("Maaf, pilihan mata uang tidak valid.")
        return

    # Mendapatkan kurs mata uang asing yang dipilih
    exchange_rate = exchange_rates[currency_choice]

    # Mendapatkan nama mata uang yang dipilih
    currency_name = {
        1: "USD",
        2: "EUR",
        3: "JPY",
        4: "GBP",
        5: "AUD",
        6: "SGD",
        7: "MYR",
        8: "IDR"
    }

    # Meminta pengguna memasukkan jumlah uang asing yang ingin ditukarkan
    amount = float(input(f"Masukkan jumlah uang {currency_name[currency_choice]} yang ingin ditukarkan: {currency_name[currency_choice]}"))

    # Mengonversi jumlah uang asing ke Rupiah
    rupiah_amount = convert_to_rupiah(amount, exchange_rate)

    print("\nProses Konversi:")
    print(f"Jumlah uang {currency_name[currency_choice]} {amount} setara dengan:")
    print(f"{currency_name[currency_choice]} {amount} * Rp{exchange_rate} = Rp{rupiah_amount}")

    # Denominasi untuk pembagian THR menggunakan konsep greedy
    denominations = [100000, 50000, 20000, 10000, 5000, 2000, 1000]

    # Input jumlah uang yang ingin ditukarkan untuk orang dewasa
    adult_amount = int(input("\nMasukkan jumlah uang yang ingin ditukarkan untuk orang dewasa: Rp"))

    # Menggunakan nilai rupiah_amount untuk penukaran uang dewasa
    adult_result, remaining_amount = calculate_change(rupiah_amount, denominations)

    # Menampilkan hasil penukaran untuk orang dewasa
    print("Hasil pembagian THR untuk orang dewasa:")
    for denom, count in adult_result.items():
        print(f"{count}x Rp{denom}")

    # Mengurangi jumlah uang rupiah dari hasil penukaran dengan jumlah rupiah yang diinput oleh pengguna
    remaining_amount = rupiah_amount - adult_amount

    # Menampilkan sisa jumlah uang setelah pembagian kepada orang dewasa
    print("Sisa jumlah uang setelah pembagian kepada orang dewasa:", remaining_amount)

    # Menggunakan sisa jumlah uang setelah pembagian kepada orang dewasa untuk penukaran THR anak-anak
    child_result, child_remaining_amount = calculate_change(remaining_amount, denominations)

    # Menampilkan hasil penukaran untuk anak-anak
    print("\nHasil pembagian THR untuk anak-anak:")
    for denom, count in child_result.items():
        print(f"{count}x Rp{denom}")

    # Menampilkan sisa jumlah uang setelah pembagian kepada anak-anak
    print("Sisa jumlah uang setelah pembagian kepada anak-anak:", child_remaining_amount)


if __name__ == "__main__":
    main()
```
